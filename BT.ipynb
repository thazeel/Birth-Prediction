```python

selected_data['was_term']=metadata['was_term']
```

```python
selected_data['group']=selected_data.index.str[0]
selected_data.head()
```

```python
train, test = train_test_split(selected_data, test_size=0.2, stratify=selected_data['group'])
train_groups = train.groupby('group')
test_groups = test.groupby('group')
```

```python
models = {}
for group, group_data in train_groups:
    X = group_data.drop(['group', 'was_term'], axis=1)
    y = group_data['was_term']
    model = RandomForestClassifier()
    model.fit(X, y)
    models[group] = model
```

```python
models = {}
for group, group_data in train_groups:
    X = group_data.drop(['group', 'was_term'], axis=1)
    y = group_data['was_term']
    model = RandomForestClassifier()
    model.fit(X, y)
    models[group] = model
```

```python
predictions = []
true_labels = []
for group, group_data in test_groups:
    model = models[group]
    X = group_data.drop(['group', 'was_term'], axis=1)
    y = group_data['was_term']
    y_pred = model.predict(X)
    predictions.extend(y_pred)
    true_labels.extend(y)
```

```python
accuracies = {}
for group, group_data in test_groups:
    y_true = true_labels[:len(group_data)]
    y_pred = predictions[:len(group_data)]
    accuracy = accuracy_score(y_true, y_pred)
    accuracies[group] = accuracy
    predictions = predictions[len(group_data):]
    true_labels = true_labels[len(group_data):]
```

```python
for group, accuracy in accuracies.items():
    print(f'Group {group}: accuracy={accuracy}')
```

```python
for group, model in models.items():
    feature_names = X.columns
    importances = model.feature_importances_
    max_importance_index = importances.argmax()
    max_importance_feature = feature_names[max_importance_index]
    print(f'Group {group}: {max_importance_feature} (importance={importances[max_importance_index]})')
```
