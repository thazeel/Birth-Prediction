```python
metadata=pd.read_csv(r"E:\BT workshop\PTB\metadata\metadata.csv")
relabd_species=pd.read_csv(r"E:\BT workshop\PTB\taxonomy\taxonomy_relabd.species.csv")
metadata.index=[i for i in metadata['specimen']]
```

```python
relabd_species.head()
```

```python
relabd_species.index=[i for i in relabd_species['specimen']]
relabd_species.drop('specimen',inplace=True,axis=1)
```

```python
relabd_species.head()
```

```python
relabd_species[relabd_species<0.001]=0
```

```python
groups = relabd_species.groupby(relabd_species.index.str[0])
result = {}
for name, group in groups:
    mask = group > 0
    col_means = mask.mean()[mask.mean() >= 0.5]
    result[name] = col_means.index.values
```

```python
result
```

```python
bacterias=set()
for key, value in result.items():
    bacterias.update(value)
```

```python
bacterias
```

```python
relab_bact=relabd_species[bacterias]
```

```python
relab_bact.to_csv('relab_bact.csv', index=False)
```
